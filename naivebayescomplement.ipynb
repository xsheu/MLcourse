```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import ComplementNB
from sklearn.metrics import accuracy_score

```

```python
# Load a subset of the dataset
categories = ['rec.autos', 'sci.space']
data = fetch_20newsgroups(subset='train', categories=categories, remove=('headers', 'footers', 'quotes'))
test_data = fetch_20newsgroups(subset='test', categories=categories, remove=('headers', 'footers', 'quotes'))

# Convert text to word counts
vectorizer = CountVectorizer(stop_words='english')
X_train = vectorizer.fit_transform(data.data)
X_test = vectorizer.transform(test_data.data)

# Train ComplementNB
model = ComplementNB()
model.fit(X_train, data.target)

# Predict and evaluate
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(test_data.target, y_pred))


from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print("Precision=",precision_score(test_data.target,y_pred))
print()
print("Recall=",recall_score(test_data.target,y_pred))
print()
print("F1 score=",f1_score(test_data.target,y_pred))
print()
print("Accuracy=",accuracy_score(test_data.target,y_pred))


```
